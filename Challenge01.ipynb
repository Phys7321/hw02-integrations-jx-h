```python
import scipy.integrate as itg
import matplotlib.pyplot as plt
import numpy as np
import math

a_array = np.logspace(-20, 2, 200)
t_array = list(map(lambda a: 2*math.sqrt(2)*itg.quad(lambda x: 1/math.sqrt(a**4 - x**4), 0, a)[0], a_array))
plt.loglog(a_array, t_array, '.')

plt.xlabel('a')
plt.ylabel('T')
plt.title('see MATLAB part for details')
plt.show()
```
